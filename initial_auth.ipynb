# Local Strava Setup 

```python

import json #Will help with saving credentials to a file
import webbrowser 
from stravalib.client import Client

#enter creditials from Strava API (deleted for privacy)
client_id = "<enter_client_id>"
client_secret = "<enter_client_secret>"

#initialize strava tool
client = Client()

#adding URL from Strava after API authorization and setting keys (scope)
redirect_url = "http://127.0.0.1:5000/authorization"
scope = ["read_all", "profile:read_all", "activity:read_all"]


#generating web address for Strava permission
auth_url = client.authorization_url(
    client_id=client_id,
    redirect_uri=redirect_url,
    scope=scope
)

#opening default browser to that Strava permission page
webbrowser.open(auth_url)
print()

#input prompt to input url opened in browser
code = input("Paste the code from the URL here: ")

#exchanging that temporary one-time code for permanent Refresh Tokens
token_response = client.exchange_code_for_token(
    client_id=client_id,
    client_secret=client_secret,
    code=code
)

#Save the token response as a JSON file
with open("tokens.json", "w") as f:
    json.dump(token_response, f)

print("Tokens saved. You will not need to do this again.")
```
